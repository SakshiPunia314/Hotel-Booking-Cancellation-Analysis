```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#loading the datasets
df = pd.read_csv('hotel_bookings.csv')
```

```python
#Exploratory Data Analysis and Data Cleaning
df.head()
```

```python
df.info()   #for datatypes
```

```python
#converting object datatype of reservation_status_date to datetime datatype
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
df.info()
```

```python
df.describe(include='object')   #describe giver integer values only..so including 'object'
```

```python
for col in df.describe(include='object').columns:
  print(col)
  print(df[col].unique())
  print('-'*50)
```

```python
df.isnull().sum()       #no. of null values in each column
```

```python
#agent and company have huge null values and delete these columns
df.drop(['company','agent'],axis=1,inplace=True,errors='ignore')
#remove all null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

```python
#adr is average daily rate or we can say price
#so 5400 is an outlier and we remove it
df=df[df['adr']<5000]
df.describe()
```

```python
#DATA ANALYSIS AND VISUALIZATION
```

```python
#amount of reservations cancelled
canceled_perc= df['is_canceled'].value_counts(normalize=True) #normalize for percentage
print(canceled_perc)
plt.figure(figsize=(4,3)) #size of plot(width and height resp.)
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.title('Reservation status count')
plt.show()
```

```python
#hotelwise status
plt.figure(figsize=(8,4))
ax=sns.countplot(x='hotel',hue='is_canceled',data=df,palette='Blues')
legend_labels,_=ax. get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('no. of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#percentages for resort hotel
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#percentages for city hotel
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#comparing the average daily rates of hotels
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#from the plot...
#the spikes represent adr of weekends and month ends is high
#Resort hotels comparativly have more price for some days
```

```python
#Monthwise status
#first extract months from date_time
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_=ax. get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('month')
plt.ylabel('no. of reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
#per month adr
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
#from the above 2 plots , we get to know that..
#Jan have highest cancelletion becos os high ADR
#Aug and Sept have highest cancelletion becos os less ADR
#so high cancellation is becoz of high ADRs
```

```python
#Countrywise status
#since we have many countries..we'll look for top 10 countries only
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(5,5))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%2f',labels=top_10_country.index)
plt.show()
```

```python
#from above plot..portugal have highest cancelation
```

```python
#percentage of From where the customers with cancellation are coming
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
#comparing ADRs of cancelled and not cancelled
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='Cancelled')
plt.legend()
plt.show()
```
